```python
import pandas as pd
import numpy as np
from faker import Faker
import random

fake = Faker()
np.random.seed(42)

N = 1000000

asset_types = ['Laptop', 'Server', 'Mobile', 'VM', 'IoT']
manufacturers = ['Dell', 'HP', 'Lenovo', 'Cisco', 'Apple', 'VMware']
os_types = ['Windows', 'Linux', 'macOS']
departments = ['Finance', 'IT', 'HR', 'Operations', 'Security']
roles = ['User', 'Admin', 'Manager', 'Auditor']

data = []

for i in range(N):

    asset_type = random.choice(asset_types)

    asset_age = np.random.randint(30, 3650)

    assigned = np.random.binomial(1, 0.85)

    software_count = np.random.poisson(40)

    unauthorized_flag = np.random.binomial(1, 0.08)

    missing_critical_software = np.random.binomial(1, 0.06)

    patch_status = random.choices(
        ['Updated', 'Pending', 'Outdated'],
        weights=[0.6, 0.25, 0.15]
    )[0]

    last_patch_days = np.random.randint(0, 300)

    vuln_count = np.random.poisson(4)
    max_cvss = round(np.random.uniform(0, 10), 2)

    critical_vuln = 1 if max_cvss > 7 else 0

    after_hours = np.random.binomial(1, 0.25)

    idle_days = np.random.randint(0, 30)

    compliance_score = max(0, min(100,
        100
        - (unauthorized_flag * 20)
        - (missing_critical_software * 15)
        - (critical_vuln * 25)
        - (after_hours * 10)
        - np.random.normal(10, 5)
    ))

    risk_score = max(0, min(100,
        (unauthorized_flag * 25)
        + (critical_vuln * 30)
        + (after_hours * 15)
        + (vuln_count * 3)
        + np.random.normal(20, 10)
    ))

    asset_risk_class = (
        'Critical' if risk_score > 80 else
        'High' if risk_score > 60 else
        'Medium' if risk_score > 40 else
        'Low'
    )

    data.append([
        fake.uuid4(),
        asset_type,
        random.choice(manufacturers),
        fake.uuid4(),
        fake.uuid4(),
        fake.date_this_decade(),
        fake.date_this_decade(),
        asset_age,

        fake.uuid4(),
        random.choice(departments),
        random.choice(roles),
        np.random.binomial(1, 0.2),
        random.choice(['Assigned','Unassigned','Retired']),

        random.choice(os_types),
        f"{random.randint(7,12)}",
        np.random.randint(2, 64),
        np.random.randint(4, 256),
        np.random.randint(128, 2000),
        np.random.binomial(1, 0.4),

        software_count,
        unauthorized_flag,
        missing_critical_software,

        patch_status,
        last_patch_days,
        vuln_count,
        max_cvss,
        critical_vuln,

        fake.ipv4(),
        random.choice(['US','UK','DE','IN','KE','RW','FR']),
        np.random.binomial(1, 0.3),
        np.random.binomial(1, 0.2),
        np.random.binomial(1, 0.1),

        round(np.random.uniform(1, 12),2),
        idle_days,
        np.random.randint(0, 24),
        after_hours,

        round(np.random.uniform(200, 5000),2),
        round(np.random.uniform(0.05, 0.2),2),
        round(np.random.uniform(100, 5000),2),
        np.random.binomial(1, 0.15),

        compliance_score,
        risk_score,
        np.random.binomial(1, 0.12),

        asset_risk_class,
        np.random.binomial(1, 0.08),
        np.random.binomial(1, 0.05)
    ])

columns = [
    'asset_id','asset_type','manufacturer','serial_number','device_id',
    'purchase_date','warranty_end_date','asset_age_days',

    'assigned_user_id','assigned_department','asset_owner_role',
    'is_shared_asset','assignment_status',

    'os_type','os_version','cpu_cores','ram_gb','disk_gb','virtualized_flag',

    'installed_software_count','unauthorized_software_flag','critical_software_missing_flag',

    'patch_status','last_patch_days','vulnerability_count','max_cvss_score','critical_vulnerability_flag',

    'last_login_ip','last_login_country','vpn_used','external_access_flag','unusual_login_pattern_flag',

    'avg_daily_active_hours','idle_days_last_30','peak_usage_hour','after_hours_usage_flag',

    'asset_value_usd','depreciation_rate','book_value_usd','replacement_due_flag',

    'compliance_score','risk_score','policy_violation_flag',

    'asset_risk_class','security_incident_flag','data_exfiltration_suspected_flag'
]

df = pd.DataFrame(data, columns=columns)

df.to_csv("it_asset_inventory_dataset.csv", index=False)

print(df.shape)
```
